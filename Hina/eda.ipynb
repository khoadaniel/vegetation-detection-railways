## Create a dictionary from image files against labels and respective pixel counts

```python
dict_of_img={}
```

```python
path_images = 'rs19_val/uint8/rs19_val/'

breaker = 0

for masks_img in os.listdir(path_images):
    #breaker +=1
    if masks_img.endswith('.png'):
        temp_mask = cv2.imread(path_images+masks_img)
        labels, count = np.unique(temp_mask[:,:,0], return_counts=True)
        dict_of_img[masks_img]=dict(zip(labels,count))
```

```python
dict_of_img
```

```python
dict_df=pd.DataFrame(dict_of_img).T
```

```python
dict_df.to_csv('image_label_dict_df.csv')
```

```python
pixels = {}
labels = {}

for key in dict_of_img:
    item = dict_of_img[key]
    for labelKey in item:
        if  labelKey in pixels:
            pixels[labelKey] = pixels[labelKey] + item[labelKey]
            labels[labelKey] = labels[labelKey] + 1
        else:
            pixels[labelKey] = item[labelKey]
            labels[labelKey] =1       
            
print(pixels)
print(labels)
```
